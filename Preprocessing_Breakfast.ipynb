```python
import os

folder_path = "./pancake_yes/"
files_and_dirs = os.listdir(folder_path)

current_person = "init"
list_of_people_and_actions = []

current_person = files_and_dirs[0][:3]
with open(folder_path+files_and_dirs[0], mode="r") as file2:
    lines = file2.readlines()
list_of_actions = []
for line in lines:
    list_of_actions.append(line.split(" ")[1])
list_of_people_and_actions.append([current_person, list_of_actions])


for file in files_and_dirs:
        if current_person == file[:3]:
            pass
        else:
            current_person = file[:3]
            with open(folder_path+file, mode="r") as file2:
                lines = file2.readlines()
            list_of_actions = []
            for line in lines:
                list_of_actions.append(line.split(" ")[1])
            list_of_people_and_actions.append([current_person, list_of_actions])
for i in list_of_people_and_actions:
    print(i)

```
